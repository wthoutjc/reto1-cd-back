```python
# IPYNB para testear consultas avanzadas del modelo
import pandas as pd
from sodapy import Socrata
```

```python
client = Socrata("www.datos.gov.co", None)
results = client.get("gt2j-8ykr", limit=1000)

df = pd.DataFrame.from_records(results)
df
```

```python
df['id_de_caso'] = df['id_de_caso'].astype(str).astype(int)
df['departamento'] = df['departamento'].astype(str).astype(int)
df['ciudad_municipio'] = df['ciudad_municipio'].astype(str).astype(int)
df['edad'] = df['edad'].astype(str).astype(int)
df['fecha_reporte_web'] = pd.to_datetime(df['fecha_reporte_web'])
df['fecha_de_notificaci_n'] = pd.to_datetime(df['fecha_de_notificaci_n'])
df['fecha_inicio_sintomas'] = pd.to_datetime(df['fecha_inicio_sintomas'])
df['fecha_diagnostico'] = pd.to_datetime(df['fecha_diagnostico'])
df['fecha_recuperado'] = pd.to_datetime(df['fecha_recuperado'])
df['fecha_muerte'] = pd.to_datetime(df['fecha_muerte'])
df.dtypes
```
